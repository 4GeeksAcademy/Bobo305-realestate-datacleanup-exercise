#### Normalize the variable of prices for each population and plot the 4 histograms in the same plot (★★★) 

```python
import pandas as pd

# Load the dataset
ds = pd.read_csv('assets/real_estate.csv', sep=';')

population_columns = ['level1', 'level2', 'level3', 'level4', 'level5', 'level6', 'level7', 'level8']
ds['population_identifier'] = ds[population_columns].apply(lambda x: '-'.join(x.astype(str)), axis=1)

unique_populations = ds['population_identifier'].unique()

most_expensive_homes = pd.DataFrame(columns=ds.columns)

for population in unique_populations:
    subset = ds[ds['population_identifier'] == population]
    most_expensive_home = subset.loc[subset['price'].idxmax()]
    most_expensive_homes = pd.concat([most_expensive_homes, most_expensive_home.to_frame().T])


most_expensive_homes.drop(columns=['population_identifier'], inplace=True)


most_expensive_homes.reset_index(drop=True, inplace=True)

print("Most Expensive Homes for Each Population")
print(most_expensive_homes)
```

```python
## HERE: plot the coordinates of the estates
import pandas as pd
ds = pd.read_csv('assets/real_estate.csv', sep=';')
from ipyleaflet import Map, basemaps, Marker, CircleMarker
south_belt_populations = ["Fuenlabrada", "Leganés", "Getafe", "Alcorcón"]

madrid_map = Map(center=(40.4168, -3.7038), zoom=11, basemap=basemaps.Stamen.Terrain)
## PUT HERE YOUR CODE:

for population in south_belt_populations:
    subset = ds[ds['level5'] == population]
    for index, row in subset.iterrows():
        lat = float(row['latitude'].replace(',', '.'))  
        lon = float(row['longitude'].replace(',', '.'))  
        
        # Define the marker color based 
        if population == "Fuenlabrada":
            color = "red"
        elif population == "Leganés":
            color = "blue"
        elif population == "Getafe":
            color = "green"
        elif population == "Alcorcón":
            color = "purple"
        
       
        marker = CircleMarker(location=(lat, lon), color=color, radius=3)
        
        
        madrid_map.add_layer(marker)

# Display the map
#this toook forever to do >.<
madrid_map
```

#### What can you say about the price per square meter between the towns of Getafe and Alcorcón?

```python
import pandas as pd
ds = pd.read_csv('assets/real_estate.csv', sep=';')
getafe_data = ds[ds['level5'] == "Getafe"]
alcorcon_data = ds[ds['level5'] == "Alcorcón"]

getafe_price_per_sqm = getafe_data['price'] / getafe_data['surface']
alcorcon_price_per_sqm = alcorcon_data['price'] / alcorcon_data['surface']

mean_price_per_sqm_getafe = getafe_price_per_sqm.mean()
mean_price_per_sqm_alcorcon = alcorcon_price_per_sqm.mean()


print("Mean Price Per Square Meter in Getafe:", mean_price_per_sqm_getafe)
print("Mean Price Per Square Meter in Alcorcón:", mean_price_per_sqm_alcorcon)
# Mean Price Per Square Meter in Getafe: 2066.314949251463
# Mean Price Per Square Meter in Alcorcón: 2239.302480199618
```

#### Reload the data, is the average of "Valdemorillo" and "Galapagar" prices the same?

#### Make the same plot for 4 different populations (level5) and rearrange them on the same graph? (★★☆) 
Hint: make a scatter plot of each population using subplots.

```python
import pandas as pd
import matplotlib.pyplot as plt
ds = pd.read_csv('assets/real_estate.csv', sep=';')

populations = ["Fuenlabrada", "Leganés", "Getafe", "Alcorcón"]  


fig, axs = plt.subplots(2, 2, figsize=(12, 8))
fig.suptitle("Scatter Plots of Surface Area vs. Price by Population", fontsize=16)

for i, population in enumerate(populations):
    row = i // 2
    col = i % 2
    ax = axs[row, col]
    
    
    population_data = ds[ds['level5'] == population]
    
 
    ax.scatter(population_data['surface'], population_data['price'], alpha=0.5)
    ax.set_title(f"Population: {population}")
    ax.set_xlabel("Surface Area (sqm)")
    ax.set_ylabel("Price")
    ax.grid(True)

plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
#unable to see plot due to codesspace error
```
